```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'D:/boston_housing.csv'
boston_data = pd.read_csv(file_path)

# Set plot style
sns.set(style="whitegrid")

# 1. Boxplot for the "Median value of owner-occupied homes" (MEDV)
plt.figure(figsize=(8, 6))
sns.boxplot(y=boston_data['MEDV'], color='skyblue')
plt.title("Boxplot of Median Value of Owner-Occupied Homes (MEDV)", fontsize=14)
plt.ylabel("MEDV ($1000s)", fontsize=12)
plt.show()
```

```python
# 2. Bar plot for the Charles River variable (CHAS)
plt.figure(figsize=(8, 6))
chas_counts = boston_data['CHAS'].value_counts()
sns.barplot(x=chas_counts.index, y=chas_counts.values, palette='viridis')
plt.title("Bar Plot of Charles River Variable (CHAS)", fontsize=14)
plt.xlabel("CHAS (1: Bounds River, 0: Does Not Bound)", fontsize=12)
plt.ylabel("Number of Towns", fontsize=12)
plt.xticks([0, 1], ['Does Not Bound', 'Bounds'], fontsize=10)
plt.show()
```

```python
# 3. Boxplot of MEDV vs. discretized AGE groups
# Discretize the AGE variable into three groups
boston_data['AGE_Group'] = pd.cut(
    boston_data['AGE'],
    bins=[0, 35, 70, 100],
    labels=['35 years and younger', 'Between 35 and 70 years', '70 years and older']
)

plt.figure(figsize=(10, 6))
sns.boxplot(x='AGE_Group', y='MEDV', data=boston_data, palette='pastel')
plt.title("Boxplot of MEDV by Age Groups", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("MEDV ($1000s)", fontsize=12)
plt.xticks(fontsize=10)
plt.show()
```

```python
# 4. Scatter plot of NOX vs. INDUS
plt.figure(figsize=(8, 6))
sns.scatterplot(x='INDUS', y='NOX', data=boston_data, color='purple')
plt.title("Scatter Plot of Nitric Oxide Concentrations (NOX) vs. Non-Retail Business Acres (INDUS)", fontsize=14)
plt.xlabel("Proportion of Non-Retail Business Acres (INDUS)", fontsize=12)
plt.ylabel("Nitric Oxide Concentrations (NOX)", fontsize=12)
plt.show()
```

```python
# 5. Histogram for the pupil-to-teacher ratio (PTRATIO)
plt.figure(figsize=(8, 6))
sns.histplot(boston_data['PTRATIO'], bins=15, kde=False, color='teal')
plt.title("Histogram of Pupil-to-Teacher Ratio (PTRATIO)", fontsize=14)
plt.xlabel("Pupil-to-Teacher Ratio", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```
